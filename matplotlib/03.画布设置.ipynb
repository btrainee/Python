```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# 画布分区

```python
x = np.linspace(0,10,100)
sin_y = np.sin(x)
plt.subplot(2,2,1) #对画布进行分区，两行两列到1区(共4区)
plt.plot(x,sin_y)
plt.xlim(-5,20) #修改x\y的坐标
plt.ylim(-2,2)
plt.subplot(2,2,4) # 画到4区
plt.plot(x,np.cos(x))
plt.show()
```

# 画布大小调整

```python
x = np.linspace(0,10,100)
sin_y = np.sin(x)
plt.plot(x, sin_y)
fig = plt.gcf()
fig.set_size_inches(12.5, 8.5)
```

# 绘制参考线

```python
x = np.linspace(0,10,100)
sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.grid(ls='--', c='darkblue') # 网格线
plt.axhline(y=0, c='red', ls='-', lw=2) # 绘制参考线
plt.axvline(x=0, c='red', ls='-', lw=2)
```

# 绘制参考区域

```python
x = np.linspace(0,10,100)
sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.grid(ls='--', c='darkblue') # 网格线
plt.axvspan(xmin=4, xmax=6, facecolor='r', alpha=0.5)
plt.axhspan(ymin=-0.25, ymax=0.25, facecolor='blue', alpha=0.8)
```
