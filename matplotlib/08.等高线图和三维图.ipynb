```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
#计算x和y的相交点
X,Y=np.meshgrid(x,y)
#计算Z的坐标
Z=np.sqrt(X**2+Y**2)
```

```python
plt.contour(X,Y,Z)
plt.show()
```

```python
plt.contourf(X,Y,Z)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #导入3D包
```

```python
X=[1,1,2,2]
Y=[3,4,4,3]
Z=[1,100,1,1]
# 创建Axes3D对象
figure=plt.figure()
ax=Axes3D(figure)
ax.plot_trisurf(X,Y,Z)
plt.show() #pycharm可显示三维信息，可旋转
```
