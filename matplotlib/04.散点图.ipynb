```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
sin_y = np.sin(x)
plt.scatter(x, sin_y, s=10, color='yellow') # 也可以用 plt.plot(x,sin_y,'o')，且速度优于scatter，但是scatter可以设置点的大小和颜色等（设置点的形式）
plt.show()
```

绘制10种大小 100种颜色的散点图

```python
np.random.seed(0) # 执行多次每次获取的随机数都是一样的
x=np.random.rand(100)
y=np.random.rand(100)
# 生成10中大小
size = np.random.rand(10)*100
#
color = np.random.rand(100)
plt.scatter(x,y,s=size,c=color,alpha=0.7) 
# s表示点大小（点大小个数可以与点个数不一致，会循环使用）,c表示点的颜色（颜色个数需要与点的个数一致）,alpha表示透明度
plt.show()
```
