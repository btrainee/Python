```python
import matplotlib.pyplot as plt
import numpy as np
```

# 普通柱状图

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
y = [1000,2000,3000,4000]
plt.bar(x,y)
plt.show()
```

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
y = [1000,2000,3000,4000]
plt.bar(x,y,width=3)
plt.xticks(x,x)
plt.show()
```

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
x_label = [str(i)+"年" for i in x]
y = [1000,2000,3000,4000]
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.bar(x,y,width=3)
plt.xticks(x, x_label)
plt.xlabel('年份',labelpad=20)
plt.ylabel('销量')
plt.title('根据年份销量对比图',pad=20)
plt.show()
```

barh进行水平方向的绘制

```python
np.random.seed(0)
x=np.arange(5)
y=np.random.randint(-5,5,5)
## 切分画布
plt.subplot(1,2,1) #1行2列
vbar=plt.bar(x,y,color='blue')
#在0位置添加蓝色的线条
plt.axhline(0,color='yellow',linewidth=2)
#对y值大于0设置为蓝色，小于0的柱设置为绿色
for bar,height in zip(vbar,y):
    if height<0:bar.set(color='green')
        
plt.subplot(1,2,2)
plt.barh(x,y,color='red') ## x与y进行了对换
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.axvline(0,color='green',linewidth=2)

plt.show()
```

# 堆积图

```python
x=[1,2,3,4]
classes=['一班','二班','三班','四班']
ch=[70,80,85,77]
english=[80,85,88,82]
#chinese=[78, 89, 88, 90]
math=[90,96,88,87]

plt.bar(x,ch,label='Chinese',color='orange')#设置图表颜色
plt.bar(x,english,bottom=ch,color='#43cd80',label='English')
cheng=[ch[i]+english[i] for i in range(len(x))]
plt.bar(x,english,bottom=cheng,color='#804dff',label='Math')

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

plt.xlabel('班级', fontsize=15)  #横坐标
plt.ylabel('成绩', fontsize=15)  #纵坐标
plt.ylim([0,300])
plt.xlim([0,6])
plt.title("学生成绩条形图", fontsize=20) #标题
plt.xticks(x,classes) #xticks() 对应坐标名称

#plt.text(1,81,70) #添加文字text(横坐标，纵坐标，文字)
for i in range(0,len(x)):
    plt.text(i+1,ch[i]-10,ch[i],color='white')
    plt.text(i+1,ch[i]+english[i]-10,english[i],color='white')
    plt.text(i+1,ch[i]+english[i]+math[i]-20,math[i],color='white')
    plt.text(i+1,ch[i]+english[i]+math[i]+5,ch[i]+english[i]+math[i],color='black')
plt.legend(loc='upper right', fontsize=10, prop='Times New Roman')#图例
plt.grid(axis='y',color='gray',linestyle='--',linewidth=1)
plt.show()
```

```python
import pandas as pd
data = pd.DataFrame({'chinese':[70,80,85,77],'english':[80,85,88,82],'math':[90,96,88,87]},index=['一班','二班','三班','四班'])
plt.bar(x = data.index.values, height = data['chinese'], color = 'orange', label = 'chinese', tick_label = data.index.to_list())
plt.bar(x = data.index.values, height = data['english'], bottom = data['chinese'], color = '#43cd80', label = 'english', tick_label = data.index.to_list())
plt.bar(x = data.index.values, height = data['math'], bottom = data['chinese'] + data['english'], color = '#804dff', label = 'math', tick_label = data.index.to_list())
plt.legend(loc='upper right', fontsize=10, prop='Times New Roman')#图例
# plt.grid(axis='y',color='gray',linestyle='--',linewidth=1)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.xlabel('班级', fontsize=15)  #横坐标
plt.ylabel('成绩', fontsize=15)  #纵坐标
plt.ylim([0,300])
plt.xlim([-1,5])
plt.title("学生成绩条形图", fontsize=20) #标题
for i in range(0,len(data.index.values)):
    plt.text(i,ch[i]-10,ch[i],color='white')
    plt.text(i,ch[i]+english[i]-10,english[i],color='white')
    plt.text(i,ch[i]+english[i]+math[i]-20,math[i],color='white')
    plt.text(i,ch[i]+english[i]+math[i]+5,ch[i]+english[i]+math[i],color='black')
```

# 堆积百分比图

```python
data = data.div(data.sum(axis=1), axis=0)
```

```python
data
```

```python
plt.bar(x = data.index.values, height = data['chinese'], color = 'orange', label = 'chinese', tick_label = data.index.to_list())
plt.bar(x = data.index.values, height = data['english'], bottom = data['chinese'], color = '#43cd80', label = 'english', tick_label = data.index.to_list())
plt.bar(x = data.index.values, height = data['math'], bottom = data['chinese'] + data['english'], color = '#804dff', label = 'math', tick_label = data.index.to_list())
# plt.grid(axis='y',color='gray',linestyle='--',linewidth=1)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.xlabel('班级', fontsize=15)  #横坐标
plt.ylabel('成绩', fontsize=15)  #纵坐标
plt.title("学生成绩条形图", fontsize=20) #标题
plt.legend(loc='upper right', fontsize=10, prop='Times New Roman', bbox_to_anchor=(1.25,0.7))#图例
```

#  分块图

```python
w=0.4
x1=[1.0,3.0,5.0,7.0]
x2=[i+w for i in x1]
x3=[i+w for i in x2]

plt.bar(x1,ch,label='Chinese',color='orange',width=w)
plt.bar(x2,english,label='English',color='blue',width=w)
plt.bar(x3,math,label='Math',color='g',width=w)

plt.legend()

plt.title("学生成绩分块图")
plt.xlabel('class')
plt.ylabel('score')
plt.xticks(x2,classes)
plt.ylim([0, 150])
plt.show()
```
