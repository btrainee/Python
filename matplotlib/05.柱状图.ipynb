```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
y = [1000,2000,3000,4000]
plt.bar(x,y)
plt.show()
```

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
y = [1000,2000,3000,4000]
plt.bar(x,y,width=3)
plt.xticks(x,x)
plt.show()
```

```python
#x,y分别表示年份和销量
x = [1980,1985,1990,1995]
x_label = [str(i)+"年" for i in x]
y = [1000,2000,3000,4000]
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.bar(x,y,width=3)
plt.xticks(x,x_label)
plt.xlabel('年份')
plt.ylabel('销量')
plt.title('根据年份销量对比图')
plt.show()
```

barh进行水平方向的绘制

```python
np.random.seed(0)
x=np.arange(5)
y=np.random.randint(-5,5,5)
## 切分画布
plt.subplot(1,2,1) #1行2列
vbar=plt.bar(x,y,color='blue')
#在0位置添加蓝色的线条
plt.axhline(0,color='yellow',linewidth=2)
#对y值大于0设置为蓝色，小于0的柱设置为绿色
for bar,height in zip(vbar,y):
    if height<0:bar.set(color='green')
        
plt.subplot(1,2,2)
plt.barh(x,y,color='red') ## x与y进行了对换
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.axvline(0,color='green',linewidth=2)

plt.show()
```
