```python
import matplotlib.pyplot as plt
```

1.绘制直线

```python
# 绘制两个点的连线(1,2) (4,8)
plt.plot([1,4],[2,8])
plt.show()
```

2.绘制折线

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

3.设置样式

```python
plt.plot(x,y,'--r',linewidth=5,label='图例')
# 添加xy轴名称
plt.xlabel('x')
plt.ylabel('y=x^2')
# 设置标题
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.title('多个点绘制折线图')
plt.legend() #默认在左上角
plt.legend(loc='upper right',fancybox=True,framealpha=0.5,shadow=True,borderpad=1)  #位置、边框、透明度、阴影、边框宽度
plt.show()
```
