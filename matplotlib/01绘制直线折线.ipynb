```python
import matplotlib.pyplot as plt
```

1.绘制直线

```python
# 绘制两个点的连线(1,2) (4,8)
plt.plot([1,4],[2,8])
plt.show()
```

2.绘制折线

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

3.设置样式

```python
plt.plot(x,y,'--r',linewidth=5,label='图例')
# 添加xy轴名称
plt.xlabel('x')
plt.ylabel('y=x^2')
# 设置标题
plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.title('多个点绘制折线图')
plt.legend() #默认在左上角
plt.legend(loc='upper right',fancybox=True,framealpha=0.5,shadow=True,borderpad=1)  #位置、边框、透明度、阴影、边框宽度
plt.show()
```

4.绘图参数

* x：x轴上的数据
* y：y轴上的数据
* ls：折线的风格（'-', ',' '--',  '-.', ': '）
* lw：线条宽度
* c：颜色
* marker：线条上点的形状
* markersize：线条上点的大小
* markeredgecolor：点的边框色
* markerfacecolor：点的填充色
* label：文本标签

```python
import numpy as np
```

```python
x = np.linspace(0, 100, 1000)
y = np.sin(x)
```

```python
plt.plot(x, y, c='red', ls=':', lw =2, marker='o', markersize=5, markeredgecolor='blue', markerfacecolor='black',\
        label="y=sin(x)")
plt.legend(loc='best')
plt.title='基本绘图参数设置'
```

```python
x = np.arange(0,1.1,0.01)
y = x**2
```

```python
plt.figure(figsize=(6.4,4.84), dpi=100, facecolor='blue')
plt.xlabel('x1轴')
plt.ylabel('y1轴')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(x,y,label='y=x^2')
#plt.rcParams['font.sans-serif']=['SimHei'] # matplotlib默认不支持中文，添加此行用来正常显示中文标签
plt.legend(loc='best')
plt.title('y=x^2')
#plt.savefig('keshihua.png')
plt.show()
```

```python
x = np.linspace(-2,2,1000)
y = np.exp(x)

plt.plot(x,y,ls='-',lw=2,color='g')

plt.title('center demo',fontdict={
                    'fontsize': 20,  #字体大小
                    'fontweight':'light',  #字体粗细, ‘light’, ‘normal’, ‘medium’, ‘semibold’, ‘bold’, ‘heavy’, ‘black’
                    'color':'r',  #字体颜色
                    'verticalalignment':'top',# 垂直对齐方式，'center', 'top', 'bottom', 'baseline', 'center_baseline'
                    'horizontalalignment':'center',#睡哦ing对齐方式，'center', 'right', 'left'
                    'family':'Times New Roman',   #字体系列，它的值有很多，比如 “宋体”，“微软雅黑”等等
                    'fontstyle':'italic'    #字体风格：它的值有：normal（默认值）、italic（斜体）、oblique（倾斜）
})

plt.show()
```
